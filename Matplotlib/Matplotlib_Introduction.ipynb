### Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> Quickly
* OO API -> Advanced

#### Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)")
plt.plot(x, np.cos(x), color="red", label="cos(x)")
plt.title("Sin(x) and Cos(x)")
plt.xlabel("x")
plt.ylabel('F(x)')

#plt.xlim([0, 4])
#plt.ylim([-0.75, 0.75])
#plt.axis([0, 4, -0.7, 0.6]) # thay the cho xlim, ylim
#plt.axis("tight")
plt.axis("equal")
plt.axis()
plt.legend()
```

#### Object-Oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x, y)

ax.set(title="A simple plot", xlabel="x-axis", ylabel="y-axis")
```

#### Line

#### Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x))
```

```python
#OO API
#Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

sizes = 1000*rng.rand(100)
colors = rng.rand(100)
fig, ax = plt.subplots()

img1 = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.3)

fig.colorbar(img1)
```

### 3. Bar
* Vertical
* Horizontal

```python
# Prepare data from a python dictionary
soft_drink_prices = {"Coke": 10, "Pessi": 12, "Sprite": 8}
soft_drink_prices.keys()
```

```python
soft_drink_prices.values()
```

```python
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price($)")
```

```python
fig, ax = plt.subplots()
#barh: bar horizontal
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
```

#### 4. Histogram

```python
# Prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
student_height[:10]
```

```python
#pyplot API
plt.hist(student_height, bins=30)
```

### Subplots

```python
# Option 1:
x = np.linspace(-5, 5, 20)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2, color="green")
ax1.plot(x, x*2, color="yellow")

ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
ax4.hist(student_height, bins=30, color="green")
```

```python
# Option 2:
x = np.linspace(-5, 5, 20)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax[0][0].plot(x, x/2, color="green")
ax[0][0].plot(x, x*2, color="yellow")

ax[0][1].scatter(np.random.random(10), np.random.random(10))
ax[1][0].bar(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
ax[1][1].hist(student_height, bins=30, color="green")
```

## Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./Data/california_cities.csv')
cities.head()

cities["area_total_km2"].sort_values(ascending=False)
```

```python
# Extract Latd = Vi do and Longtd = Kinh Do
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
#Plot using Pyplot API
plt.scatter(lon, lat, 
            c=np.log10(population), 
            cmap='viridis',
            s=area, 
            linewidths=0,
            alpha=0.5
            )

plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')


plt.colorbar(label='log$_{10}${population}')
plt.clim(3,7)

#Create a legend for cities size
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area,label=str(area) + 'km$^2$', c='b', alpha=0.4)
plt.legend(labelspacing=1, scatterpoints=1, title='City Area')

plt.title('California Cities: Populartion and Area Distribution')
```
